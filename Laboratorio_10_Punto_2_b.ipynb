np.random.seed(19680801) # Fijar la semilla para que los números aleatorios sean reproducibles
# Función para dibujar una imagen con degradado
defimagen_degradado(ax, direccion=0.3, rango_cmap=(0, 1), extent=None, **kwargs):
phi = direccion* np.pi/ 2 # Convertir la dirección en ángulo en radianes
v = np.array([np.cos(phi), np.sin(phi)]) # Vector direccional
X = np.array([[v @ [1, 0], v @ [1, 1]],
[v @ [0, 0], v @ [0, 1]]]) # Matriz para la imagen de degradado
a, b = rango_cmap# Descomponer el rango del mapa de colores
X = a + (b -a) / X.max() * X # Ajustar los valores del degradado
im= ax.imshow(X, interpolation='bicubic', clim=(0, 1),
aspect='auto', extent=extent, **kwargs) # Mostrar la imagen con degradado
returnim
# Función para crear las barras con degradado
defbarra_degradado(ax, x, y, ancho=0.5, fondo=0):
forizquierda, arriba in zip(x, y):
derecha = izquierda + ancho
imagen_degradado(ax, extent=(izquierda, derecha, fondo, arriba), #removed extra indent
cmap=plt.cm.Blues_r, rango_cmap=(0, 0.8))
# Crear la figura y los ejes
fig, ax= plt.subplots()
ax.set(xlim=(0, 10), ylim=(0, 1)) # Establecer los límites de los ejes
# Fondo con degradado
imagen_degradado(ax, direccion=1, extent=(0, 1, 0, 1), transform=ax.transAxes,
cmap=plt.cm.RdYlGn, rango_cmap=(0.2, 0.8), alpha=0.5)
# Datos para las barras
N = 10
x = np.arange(N) + 0.15 # Posiciones de las barras
y = np.random.rand(N) # Alturas aleatorias de las barras
barra_degradado(ax, x, y, ancho=0.7) # Dibujar las barras con degradado
# Mostrar el gráfico

```python
import pandas as pd
import numpy as np

# Datos iniciales
data = {
    "Nombre": ["Juan", "Maria", "Pedro", "Ana", "Luisa", "Carlos", "Juan"],
    "Edad": [25, 30, np.nan, 29, 28, 200, 25],
    "Email": ["juan@mail.com", "maria@mail.com", "pedro@mail.com", "ana@mail.com", "luisa@mail.com", "carlos@mail.com", "juan@mail.com"],
    "Pais": ["colombia", "Colombia", "Mexico", "Argentina", "Colombia", "Brazil", "colombia"],
    "Genero": ["M", "F", "H", "F", "Femenino", "M", "M"],
    "Visitas": [5, 7, 6, 4, 8, 2, 5]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# 1. Rellenar valores faltantes
# Rellenar Edad faltante con la media de edades (excluyendo valores extremos)
edad_media = df['Edad'][(df['Edad'] < 100)].mean()
df['Edad'].fillna(edad_media, inplace=True)

# 2. Eliminar duplicados
df.drop_duplicates(inplace=True)

# 3. Estandarizar valores
# Estandarizar género
df['Genero'] = df['Genero'].replace({"M": "Masculino", "H": "Masculino", "F": "Femenino", "Femenino": "Femenino"})

# Estandarizar nombres de países
df['Pais'] = df['Pais'].str.capitalize().replace({"Brazil": "Brasil", "Mexico": "México"})

# 4. Corregir valores inválidos
# Limitar edades a un rango razonable
df['Edad'] = df['Edad'].apply(lambda x: edad_media if x >= 100 else x)

# Mostrar el DataFrame limpio
print(df)


```
