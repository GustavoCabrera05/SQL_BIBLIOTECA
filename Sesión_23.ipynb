```python
import matplotlib.pyplot as plt
import numpy as np
# Fixingrandomstatefor reproducibility
np.random.seed(19680801)
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors= np.random.rand(N)
area= (30 * np.random.rand(N))**2 # 0 to15 pointradii
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# Configurar el estilo de la visualizacin
sns.set(style="whitegrid") # Estilo de fondo blanco con grid
plt.rcParams.update({
  'axes.titlesize': 14, # Tamao de los ttulos de los ejes
  'axes.labelsize': 12, # Tamao de las etiquetas
  'xtick.labelsize': 10, # Tamao de las etiquetas del eje x
  'ytick.labelsize': 10, # Tamao de las etiquetas del eje y
  'figure.titlesize': 16, # Tamao del ttulo de la figura
  'legend.fontsize': 10, # Tamao de la leyenda
  'axes.facecolor': '#ffffff',}) # Color de fondo de los grficos (blanco)})
# Datos de ejemplo
data = sns.load_dataset("iris")
# Crear subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
# Asegurar que el fondo de la figura sea completamente blanco
fig.patch.set_facecolor('white')
# Ttulo general de la figura
fig.suptitle("Visualizacin de datos de Iris", fontsize=18) # Ttulo general
# Grfico de caja con personalizacin
sns.boxplot(x="species", y="sepal_length", data=data, palette="Set2", ax=axes[0, 0])
axes[0, 0].set_title("Boxplot: Largo del spalo por especie")
axes[0, 0].set_xlabel("Especies")
axes[0, 0].set_ylabel("Largo del spalo")
# Histograma con KDE y personalizacin
sns.histplot(data["sepal_length"], kde=True, color='steelblue', ax=axes[0, 1])
axes[0, 1].set_title("Histograma con KDE: Largo del spalo")
axes[0, 1].set_xlabel("Largo del spalo")
axes[0, 1].set_ylabel("Frecuencia")
# Grfico de dispersin con personalizacin
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", palette="coolwarm", s=100, data=data, ax=axes[1, 0])
axes[1, 0].set_title("Scatterplot: Largo vs. Ancho del spalo")
axes[1, 0].set_xlabel("Largo del spalo")
axes[1, 0].set_ylabel("Ancho del spalo")
axes[1, 0].legend(title="Especies", loc='upper right')
# Grfico de violn con personalizacin
sns.violinplot(x="species", y="petal_length", data=data, palette="Set1", inner="quartile", ax=axes[1, 1])
axes[1, 1].set_title("Violinplot: Largo del ptalo por especie")
axes[1, 1].set_xlabel("Especies")
axes[1, 1].set_ylabel("Largo del ptalo")
# Ajustar el espaciado
plt.tight_layout(rect=[0, 0, 1, 0.95]) # Deja espacio para el ttulo superior
# Mostrar la figura con el fondo blanco
plt.show()
```
